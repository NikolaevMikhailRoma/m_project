```python
import pandas as pd
import numpy as np

df_cps = pd.read_excel ('Library_and_other/для координат/Том VIII Приложение И_ЦПС.xlsx', sheet_name='ЦПС Весна 2021')
# df_cps = pd.concat(df_cps, axis=0, ignore_index=True)
list_of_wells = df_cps['Unnamed: 3'].dropna().unique().tolist()[1:]
df_cps = df_cps[df_cps['Unnamed: 3'].isin(list_of_wells)]
df_cps = df_cps[['Unnamed: 3', 'Unnamed: 40']]
# df = pd.DataFrame(df_cps['Unnamed: 3'].dropna().unique().tolist()[1:], columns = ['Скважина'])
# df
df_cps.rename(columns={"Unnamed: 3": "Скважина", "Unnamed: 40": "Средняя температура"}, inplace = True)
# df_cps['Скважина'].replace({'ТС':''}, inplace=True)
# df_cps['Скважина'] = df_cps['Скважина'].str.replace('ТС', '', regex = True)
df_cps[0:15]
```

```python
df_coord = pd.read_excel('Library_and_other/для координат/Каталог координат ТС в МСК89.xls', sheet_name='ЦПС')
# df_coord[df_coord['Название ТС'].str.contains('221')]
df_coord.rename(columns={"Название ТС": "Скважина"}, inplace = True)
df_coord['Скважина'] = 'ТС' + df_coord['Скважина']
df_coord['Скважина'] = df_coord['Скважина'].str.replace('\\', '/', regex = True)
df_coord['Скважина'] = df_coord['Скважина'].str.replace('/....', '', regex = True)
df_coord['Скважина'] = df_coord['Скважина'].str.replace('.', '/', regex = True)

df_coord = df_coord[['Скважина', 'X', 'Y']]
df_coord
```

```python
df = df_cps.merge(df_coord, how='left', on='Скважина')
df = df[df['X'].notna()]
df.rename(columns = {'Скважина' : 'Name', 'Средняя температура': 'Тср'}, inplace = True)
df = df[['Name', 'X', 'Y', 'Тср']]
df
# df = df[['']]
df.to_excel('Library_and_other/для координат/Для карты ЦПС.xlsx',
            index=False)
```
